# EDA and Feature Engineering Flight Price Prediction

## Features

The various features of the cleaned dataset are explained below:
1. Airline : The name of the airline company is stored in the airline column. Categorical feature having 6 
            different airlines.
2.Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
3. Source city: City from which the flight takes off. Categorical feature having 6 unique cities.
4. Departure time: This is a derived categorical feature obtained by grouping the time period into bins.
            It stores the info about the departure and have 6 unique time labels.
5. Stops: A categorical feature with 3 distinct values that stores the number of stops between source and destination
        CITIES.
6. Arrival Time: This is a derived categorical feature created by grouping the time interval into bins.
            It has 6 distinct time labels keep information about the arrival time.
7. Destination City: City where the flight will land. It is categorical feature having 6 unique cities.
8. Class: A categorical feature that contains information on seat class; it has 2 distinct values: Business and economical.
9. Duration: A continuous feature that displays the overall amount of time it takes to travel between the cities in hours.
10. Days left: This is a derived characteristics that is calculated by subtracting the trip date by the booking date.
11. Price: Target variable stores information of the ticket price.
